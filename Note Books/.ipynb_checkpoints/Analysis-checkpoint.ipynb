```python


import matplotlib.pyplot as plt

```

```python
top_cities = df['Location'].value_counts().head(10)
plt.figure(figsize=(10,5))
top_cities.plot(kind='bar')
plt.title("Top Hiring Cities")
plt.xlabel("City")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.savefig(r"C:\Users\Rohitpal\OneDrive\Desktop\job-market-analyzer\Images\Top Hiring Cities")

plt.show()
```

```python



# now analyzing skills 

```

```python
df.columns
```

```python
df["Skills/Description"].head()
```

```python

skills_text = df['Skills/Description'].str.lower()  # help to match properly 
```

```python
skills_text
```

```python
# COUNT IMPORTANT SKILLS THAT MOST JOB DESCRIPTION HAS ACCORDINGH TO JOB OF DATA SCIENCE -->> 

python_count = skills_text.str.contains('python').sum()

sql_count = skills_text.str.contains('sql').sum()

powerbi_count = skills_text.str.contains('power bi').sum()

ml_count = skills_text.str.contains('machine learning').sum()

excel_count = skills_text.str.contains('excel').sum()
```

```python
print("Python:", python_count)
print("SQL:", sql_count)
print("Power BI:", powerbi_count)
print("Machine Learning:", ml_count)
print("Excel:", excel_count)  
```

```python
# from above counting skills for Data science job :-- Python: 3202  
# from above counting skills for Data science job :--  SQL: 2945
# from above counting skills for Data science job :--  Power BI: 557
# from above counting skills for Data science job :--  Machine Learning: 1794
# from above counting skills for Data science job :--  Excel: 755


```

```python


# NOW VISUALIZATION OF ABOVE SKILLS 



```

```python
skills = ['Python', 'SQL', 'Power BI', 'Machine Learning', 'Excel']
counts = [
    python_count,
    sql_count,
    powerbi_count,
    ml_count,
    excel_count
]

plt.figure(figsize=(8,5))
plt.bar(skills, counts)
plt.title("Most Demanded Skills")
plt.xlabel("Skills")
plt.ylabel("Job Count")
plt.tight_layout()
plt.savefig(r"C:\Users\Rohitpal\OneDrive\Desktop\job-market-analyzer\Images\Most Demanded Skills")

plt.show()
```
